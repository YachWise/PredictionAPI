```python
import cortex

cx_local = cortex.client("local")

api_spec = {
  "name": "text-generator",
  "kind": "RealtimeAPI",
  "predictor": {
    "type": "python",
    "path": "predictor.py"
  }
}

cx_local.deploy(api_spec, project_dir=".", wait=True)
```

```python
import requests

endpoint = cx_local.get_api("text-generator")["endpoint"]
payload = {"text": "Sure do like"}
print(requests.post(endpoint, payload).text)
```
